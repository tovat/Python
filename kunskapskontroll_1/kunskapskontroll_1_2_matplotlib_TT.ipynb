### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
#Object-oriented approach: 

x = np.linspace(-4, 4)
y = x**2

#Skapar figuren och definierar axlarna
fig, ax = plt.subplots()
ax.plot(x, y)

#Lägger till en rubrik till figuren och rubriker till x- och y-axlarna
ax.set_title('Plot of y = x^2, created with the object-oriented approach')
ax.set_xlabel('Label of x-axis')
ax.set_ylabel('Label of y-axis')

plt.show()
```

```python
#Pyplot approach: 

x = np.linspace(-4, 4)
y = x**2

#Skapar figuren och definierar axlarna
plt.plot(x, y)

#Lägger till en rubrik till figuren och rubriker till x- och y-axlarna
plt.title('Plot of y = x^2, created with the Pyplot approach')
plt.xlabel('Label of x-axis')
plt.ylabel('Label of y-axis')

plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
#Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

#First subplot (scatter plot)
ax1.scatter(random_data_x, random_data_y, alpha=0.5)
ax1.set_title('Scatter plot')
ax1.set_xlabel('X-label')
ax1.set_ylabel ('Y-label')

ax2.bar(names, values)
ax2.set_title('Bar plot')
ax2.set_xlabel('Fruit')
ax2.set_ylabel('Amount')
```
